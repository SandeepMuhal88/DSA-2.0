```python
arr=[10,20,30,40,50]
print("Original array:",arr)
```

```python
# Traversing an Array
for i in range(len(arr)):
    print(f"Element at index {i} is {arr[i]}")

for i in arr:
    print(i)
# ⏱️ Time Complexity: O(n)
```

inserting element in array

```python
arr.append(60)
print("Array after insertion:", arr)
```

Insert at specific index

```python
arr.insert(2,25)
print("Array after inserting 25 at index 2:", arr)
```

Deleting element in array


```python
arr.remove(30)

print("Array after deleting 30:", arr)
```

```python
arr.pop(2)
print("Array after popping element at index 2:", arr)
```

```python
for i in arr:
    if i == 40:
        print("Found")
```

```python
if 40 in arr:
    print("Found")
```

```python
arr.sort()        # ascending
arr.sort(reverse=True)
print("Sorted array in descending order:", arr)
```

```python
arr.reverse()
print(arr)
```

```python
arr[::-1]
```
